2) List comprehensions to return multiple lists

__Example__

Make a  list of lists, called matrix, row zero is 1,1,1,1, row 1 is the numbers 1,2,3,4,5, then the next row is
1^2, 2^2, 3^2, 4^2, 5^2, up to 1^4,2^4,3^4,4^4,5^4

Long way

```python
matrix=[]
for i in range (5):
    row=[]
    for j in range(1,6):
        row.append(j**i)
    matrix.append(row)
    
matrix
```

Medium Length way

```python
matrix=[]
for i in range (5):
    row=[j**i for j in range(1,6)]
    matrix.append(row)
    
matrix
```
