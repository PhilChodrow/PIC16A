```python

class Vector:
    """
    2D vector class supporting additon, subtraction, and scalar multiplication
    """

    
    def __init__(self,x,y):
        self.x=x
        self.y=y
    
    def scalar_multiply(self,c):
        """returns a Vector with components multiplied by c"""
        return Vector(c*self.x,c*self.y)
    
    #These are self explanatory (as is the intializer)
    def __str__(self):
        return("<"+str(self.x)+","+str(self.y)+">")
    
    def __add__(self,other):
        if other.x=="nan" or self.x=="nan":
            answer_x="nan"
        else:
            answer_x=self.x+other.x
        if other.y=="nan" or self.y=="nan":
            answer_y="nan"
        else:
            answer_y=self.y+other.y
        return Vector(answer_x,answer_y)
              
    def __sub__(self,other):
        if other.x=="nan" or self.x=="nan":
            answer_x="nan"
        else:
            answer_x=self.x-other.x
        if other.y=="nan" or self.y=="nan":
            answer_y="nan"
        else:
            answer_y=self.y-other.y
        return Vector(answer_x,answer_y)
    
    def __mul__(self,other):
        if other.x=="nan" or self.x=="nan":
            answer_x="nan"
        else:
            answer_x=self.x*other.x
        if other.y=="nan" or self.y=="nan":
            answer_y="nan"
        else:
            answer_y=self.y*other.y
        return Vector(answer_x,answer_y)
    
    
    def __truediv__(self, other):
        if other.x==0:
            answer_x="nan"
        else:
            answer_x=self.x/other.x
        if other.y==0:
            answer_y="nan"
        else:
            answer_y=self.y/other.y
        return(Vector(answer_x,answer_y))
    
    def dot(self,other):
         return (self*other).x+(self*other).y    
    
    def magnitude(self):
        return (self.x**2 + self.y**2)**.5
    
    def mean(self):
        return((self.x+self.y)/2)
    
    def apply(self,f):
        return(Vector(f(self.x),f(self.y)))
```
