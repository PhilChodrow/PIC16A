```python
# (A) This works, i has value 0
# (B) This works, i has value 10
# (C) This errors, i is not defined
# (D) This errors, can't use variables previously 
#       defined in a while loop
# (E) This runs in an infinite loop

i = 0
while i in range(10):
    print(i)
```
